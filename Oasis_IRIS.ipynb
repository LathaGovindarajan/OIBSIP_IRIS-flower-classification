```python
from sklearn import datasets
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
iris=datasets.load_iris()
iris
```

```python
iris.keys()
```

```python
#converting the dataset to pandas dataframe
iris=pd.DataFrame(data=np.c_[iris['data'],iris['target']],columns=iris['feature_names']+['target'])
```

```python
iris
```

```python
species=[]
for i in range(len(iris['target'])):
    if iris['target'][i]==0:
        species.append('Setosa')
    elif iris['target'][i]==1:
        species.append('Versicolor')
    else:
        species.append('Virginica')
```

```python
iris['species']=species#adding species into data
```

```python
# to check if the 3 targets are equal
iris.groupby('species').size()
```

```python
iris.describe()#statistical analysis
```

```python
iris.isnull().sum()
```

```python
sns.countplot(data=iris,x='species')
plt.title('SPECIES')
plt.ylabel('Species count')
plt.xlabel('Species names')
plt.yticks(fontsize=13)
plt.xticks(fontsize=12,rotation=360)
plt.grid(True)
plt.box(False)
plt.show()
```

```python
from sklearn.model_selection import train_test_split
X=iris.drop(['target','species'],axis=1)
y=iris['target']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=50)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
# Create SVM classifier with radial basis function (RBF) kernel
svm = SVC(kernel='rbf')
```

```python
# Fit classifier to training data
svm.fit(X_train, y_train)
```

```python
# Predict classes of testing data
y_pred = svm.predict(X_test)
```

```python
# Calculate accuracy score of predictions
accuracy = accuracy_score(y_test, y_pred)
```

```python
print(f"Accuracy score: {accuracy}")
```
